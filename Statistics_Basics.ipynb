###3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

In statistics, dispersion is a measure of how spread out or squeezed a distribution is.


Variance and standard deviation are two common measures of dispersion that show how close data values are to the mean of a data set:

Variance


The mean squared difference between each data point and the mean. Variance is often used because it takes into account the spread of all data points. However, squaring the numbers can amplify the effect of large differences and skew the data.

Standard deviation

The square root of the variance. It's also known as the root mean square deviation. A small standard deviation means the data values are close to the mean, while a large standard deviation means the data values are spread out.

###4. What is a box plot, and what can it tell you about the distribution of data?

A **box plot** (or **box-and-whisker plot**) is a chart that shows the distribution of a data set. It displays key values like the **minimum**, **first quartile (Q1)**, **median**, **third quartile (Q3)**, and **maximum**.

- The **box** represents the middle 50% of the data (from Q1 to Q3) with a line at the **median**.
- The **whiskers** extend from the box to show the range of the data, excluding outliers.
- **Outliers** are values outside the whiskers, often shown as dots.

### What it tells you:
- **Central tendency**: The median shows the middle value.
- **Spread**: The size of the box (IQR) and whiskers show how spread out the data is.
- **Skewness**: If the median is closer to one side of the box, the data is skewed.
- **Outliers**: Any points outside the whiskers are outliers.

In short, box plots give a clear picture of the dataâ€™s spread, central value, and whether there are any unusual outliers. They're especially useful for comparing different groups or datasets.
