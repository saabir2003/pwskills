### 1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def even (l):
  sum=0
  for i in l:
    if i%2==0:
      sum+=i
  return sum
print("sum of even number in list=",even([1,2,3,4,5]))
```

### 2)Create a Python function that accepts a string and returns the reverse of that string.

```python
def rev(s):
  return s[::-1]
print("reverse of string is =",rev("hello"))
```

### 3) Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def sq(li):
  return [i**2 for i in li]
print("square of list is =",sq([1,2,3,4,5]))
```

### 4) Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def isprime(n):
    if n==1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True
print("1 is prime=",isprime(1))
print("2 is prime=",isprime(2))
print("3 is prime=",isprime(3))
print("4 is prime=",isprime(4))
```

### 5)  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fib(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
for i in fib(10):
  print(i)
```

### 6)  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power(n):
  for i in range(n):
    yield 2**i
for i in power(10):
  print(i)
```

### 7) Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    """Reads a file line by line and yields each line as a string.

    Args:
        file_path: The path to the file to read.

    Yields:
        Each line of the file as a string.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove trailing newline character

# Example usage:
file_path = '/content/Study Links.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)  # To see the output, run the code.
```

### 10)Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, string))
print(remove_vowels("hello world"))
```

### 11)Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

```python

orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]
def calculate_total(order):
    order_number, _, quantity, price_per_item = order
    total_price = quantity * price_per_item
    if total_price < 100:
        total_price += 10
    return order_number, total_price

result = [calculate_total(order) for order in orders]

print(result)
```
