```python
ohn2.drop(columns=['x','y','z'],inplace=True)
ohn2.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(ohn2.drop(columns=["price"]),ohn2.price,test_size=0.2)
```

```python

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
print(X_train.columns)
```

```python
from sklearn.linear_model import LinearRegression
mlr1=LinearRegression()
mlr1.fit(X_train,y_train)

```

```python
y_pred=mlr1.predict(X_test)
error=y_test-y_pred
print(error)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
mae=mean_absolute_error(y_pred,y_test)
mse=mean_squared_error(y_pred,y_test)
rmse=root_mean_squared_error(y_pred,y_test)
r2=r2_score(y_test,y_pred)
print(mae,mse,rmse,r2)
```

```python
def adjusted_r2_score(y_true, y_pred, n_features):
    r2 = r2_score(y_true, y_pred)
    n = len(y_true)
    return 1 - (1 - r2) * (n - 1) / (n - n_features - 1)
print(adjusted_r2_score(y_test, y_pred, n_features=12))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred, error)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted values")
plt.ylabel("error")
plt.title("error vs. Predicted Values")
plt.show()
```

# 15. Write a Python script that generates synthetic data for a polynomial relationship (degree 4), fits a polynomial regression model, and plots the regression curve

```python
np.random.seed(42)
X = np.sort(2 * np.random.rand(100, 1) - 1, axis=0)  # Values between -1 and 1
y_true = 3 * X**4 - 2 * X**3 + X**2 - 4 * X + 1
y = y_true.ravel() + np.random.normal(0, 0.2, size=X.shape[0])  # Add noise
```

```python
degree = 4
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X)


model = LinearRegression()
model.fit(X_poly, y)


X_fit = np.linspace(-1, 1, 200).reshape(-1, 1)
X_fit_poly = poly.transform(X_fit)
y_fit = model.predict(X_fit_poly)


plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data (with noise)', alpha=0.6)
plt.plot(X_fit, y_fit, color='red', label='Polynomial Regression (Degree 4)', linewidth=2)
plt.title("Polynomial Regression (Degree 4)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
